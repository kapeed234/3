```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(data=df, x='Engagement_Level', y='Progress (%)', palette='viridis')
plt.title('Progress by Engagement Level')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Time_Spent_Min', y='Clicks', hue='Engagement_Level', palette='Dark2')
plt.title('Time Spent vs Clicks by Engagement Level')
plt.show()
```
